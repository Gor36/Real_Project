В следующем модуле вас ждёт аттестация по разделу «Математика в ML».

Вам предстоит выполнить 44 тестовых задания, которые система будет проверять автоматически, так что вы сразу сможете узнать, верно ли ответили на вопрос. Задания (за редким исключением) снабжены пояснениями, чтобы вы смогли дополнительно убедиться в правильности своего ответа или выяснить, почему ваш ответ неверен.

В аттестации представлены задания по следующим темам:

* линейная алгебра и линейные модели;
* математический анализ и методы оптимизации;
* теория вероятностей;
* математика для:
    * деревьев решений;
    * ансамблей;
    * методов кластеризации;
    * методов снижения размерности.

**Очень важно!**
* На прохождение аттестации вам даётся 4 часа с момента, как вы приступили к заданиям.
* Аттестация условно разделена на две части — рекомендуем вам заложить примерно по 2 часа на каждую из них. Выполнить обе части необходимо за один четырёхчасовой подход.
* По истечении четырёх часов отправить ответы уже не получится.
* На выполнение каждого задания даётся одна попытка.
Уверены, вам будет несложно.

Повторите темы, в которых, как вы считаете, у вас есть пробелы, и приступайте к аттестации.

**Внимание!**
Минимальный порог прохождения аттестации — 50 %.

* Если вы набрали от 50 до 75 %, вы показали хороший уровень владения темой.
* Если ваш результат превышает 75 %, вы отлично усвоили все материалы темы.

**Что делать, если результат меньше 50 %?**

В таком случае аттестация будет считаться непройденной. Однако не отчаивайтесь — у вас есть ещё 2 попытки, чтобы исправить результат.

Чтобы воспользоваться следующей попыткой, напишите запрос в Support — ваша предыдущая попытка прохождения аттестации будет сброшена.

Студентам, которые не смогли преодолеть порог в 50 % за 3 попытки, будет предложена консультация с тьютором для обсуждения дальнейшей работы над курсом.
